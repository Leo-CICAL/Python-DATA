# Comprehension list juste avec  un if => if à la fin

```python
# [f(x) for x in sequence if condition]

liste = [_ for _ in range(n) if _%2==0]
liste
```

# Comprehension list avec if else => if else au milieu

```python
# [f(x)  if condition esle g(x) for x in sequence]

liste = [_ if _%2==0 else "impair" for _ in range (n)]
liste
```

# Transmitting arbitrary named parameters to a function

```python
# FIRST LET'S CREATE A RANDOM DATAFRAME

import random
import pandas as pd
eleves=['William','Hugh','Marvin','Stella','Serge','Kimiko']

liste_eleves = [random.choice(eleves) for _ in range(60)]

def note():
    return(random.randint(1,20))

notes = [note() for _ in range (60)]
notes

eleves_notes = pd.DataFrame()
eleves_notes['eleve']=liste_eleves
eleves_notes['note']=notes
eleves_notes
```

# TRANSMITTING RANK AS PARAMETER WHEN USING PANDAS APPLY()

```python
#Last time, we did a function pour avoir le top 5
# Mais 5 était static, on pouvait pas le changer 

def get_top5(df):
    return df.sort_values(by="note")[-5:]

eleves_notes.groupby('eleve').apply(get_top5)
```

# ON VA UTILISER **kwds dans la fonction apply

```python
#Difference entre named parameters et ordered parameters : def div(a,b) LA C'EST DES ORDERED, def div (a=10,b=5) LA A ET B
# SONT NOMMES 

#EN gros **kwags crée un dict avec  tous les named paramters, avec comme clé le nom de paramtres et comme valuers bah les
# valeurs qu'on va entrer

def get_top(df,**kwags):
    return df.sort_values(by="note")[-kwags['top']:]

eleves_notes.groupby('eleve').apply(get_top, top=2)
```

```python
# Le nom kwags n'est pas obligatoire ; c'est les  ** qui importent, ça veut dire "Paque moi tt ca dans un dico"

def get_top(df,**yo):
    return df.sort_values(by="note")[-yo['top']:]

eleves_notes.groupby('eleve').apply(get_top, top=2)
```

```python
# Mais on aurait pu faire trql :

def get_top(df,top,sort):
    return df.sort_values(by=sort)[-top:]

eleves_notes.groupby('eleve').apply(get_top,sort="note", top=2)

# **kwags sert quand on connait pas vraiment la list des param, mais bon c'est rare ptdr
#Mieux vaut utiliser la méthode dans cette case  
```
